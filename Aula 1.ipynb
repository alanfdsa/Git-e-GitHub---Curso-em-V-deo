# Git ≠ Github

(No texto abaixo tentamos utilizar definições que abarquem de maneira sucinta e direta o que cada um dos termos acima são. Porém, em aspectos mais avançados pode ser que as duas definições se confundam.)

## Git - Software de controle de versão (Versionamento)

É importante pois na vida de programador é usual lidarmos com várias versões do mesmo programa. Então ter em mente o processo da construção do programa é sempre útil.

Uma representação pictórica simples de versionamento de código é o seguinte:



Mas trabalhar dessa forma não é simples, já que navegar por arquivos zipados pode se tornar exponencialmente dificil.

Um software de versionamento de códigos basicamente se organiza sozinho no acompanhamento das atualizações dos programas. Sendo assim, é como se fosse uma espécie de máquina do tempo para os seus programa, que nos oferece, por exemplo, a possibilidade de saber *exatamente* o que mudou de versão para versão.

## GitHub - Plataforma de rede social para programadores



Existem muitas outras opções, como por exemplo o GitLab, Bitbucket, Phabricator, Gogs e etc...

Além disso, é importante ter em mente que o GitHub, no esquema de versionamento distribuido, se encontra na parte do repositório remoto.
